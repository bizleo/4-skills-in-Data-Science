```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as soup
```

```python
html = requests.get('https://simple.wikipedia.org/wiki/List_of_countries')
html.status_code
```

```python
bs = soup(html.content, 'html.parser')
print(bs)
```

```python
bst = bs.find(id='table')
print(bst)
```

```python
My_table = bs.find('table', {'class' : 'sortable wikitable'})
print(My_table)
```

```python
links = My_table.findAll('a')
links
```

```python
Countries = []
for link in links:
    Countries.append(link.get('title'))
    
```

```python
Countries.head()
```

```python
table = pd.DataFrame()
table['Countries'] = Countries
```

```python
table.describe()
```

```python
table = table.drop_duplicates()
```

```python
table = table.drop(1)
```

```python
table
```

```python
clean ={country.replace(' (not yet started)', '') for country in table['Countries']}
clean
```
